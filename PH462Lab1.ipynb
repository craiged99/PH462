```python
#Task 1
#First Generator

import numpy as np
import random
import matplotlib.pyplot as plt

ran = np.linspace(0,1,1000000)

xn = []
for i in range(5000):
    xn.append(random.choice(ran))
    
xn10=[]
for i in range(4990):
    xn10.append(xn[i+10])
    
xn = xn[:4990]



plt.plot(xn,xn10,marker='+',markersize=1,linestyle='None')
plt.xlabel('$X_n$')
plt.ylabel('$X_{n+10}$')
plt.title('Scatter plot using $\it{random}$ package for 5000 points')
fig = plt.gcf()

fig.savefig('Q1.png', dpi=100)
```

```python
#Task 2

import random
import numpy as np
import matplotlib.pyplot as plt

n = np.linspace(1,200,200)
ran = np.linspace(1,200,200)
n1 = []
n = list(n)
n1 = list(n1)

hund = [100]*5000

t = np.linspace(1,5000,5000)

for i in range(len(t)):
    randomno = random.choice(ran)
    time = t[i]
    
    if randomno in n:
        n.remove(randomno)
        n1.append(randomno)
    
    else:
        n1.remove(randomno)
        n.append(randomno)
        
    
    plt.plot(time,len(n),marker='+',color='black',markersize='0.8')

    

plt.plot(t,hund,linestyle='-',linewidth='2')
plt.grid()
plt.ylim([0,200])
plt.xlim([0,5000])
plt.xlabel('Time step')
plt.ylabel('No. of particles at one side')
plt.title('Time evolution of box with 200 particles')
fig = plt.gcf()

fig.savefig('Q2.png', dpi=100)
print(len(n))
print(len(n1))

```

```python
#Task 4
#75/25 Probability 


import random
import numpy as np
import matplotlib.pyplot as plt

n = np.linspace(1,200,200)
m = n
ran = np.linspace(1,2,2)
n1 = []
n = list(n)
n1 = list(n1)

hund = [100]*5000

t = np.linspace(1,5000,5000)

for i in range(len(t)):
    randomno = np.random.choice(ran,p=[0.75,0.25])
    part = np.random.choice(m)
    time = t[i]
    
    if randomno == 1:
        if part in n:
            n.remove(part)
            n1.append(part)
    
    if randomno == 2 and len(n1)>0:
        if part in n1:
            n1.remove(part)
            n.append(part)
        
    
    plt.plot(time,len(n),marker='+',color='black',markersize='0.8')
    plt.plot(time,len(n1),marker='+',color='crimson',markersize='0.8')

plt.plot(t,hund,linestyle='-',linewidth='2')
plt.ylim([0,200])
plt.xlim([0,5000])
plt.grid()
plt.xlabel('Time step')
plt.ylabel('No. of particles at one side')
plt.title('Time evolution of box - Equal probabilty') 
plt.legend(['Side 1','Side 2'])
fig = plt.gcf()

fig.savefig('Q2.png', dpi=100)
```

```python
#Task 4
#75/25 & 55/45 Probability (Same plot)

import random
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

n = np.linspace(1,200,200)
m = n
ran = np.linspace(1,2,2)
n1 = []
n = list(n)
n1 = list(n1)

hund = [100]*5000

t = np.linspace(1,5000,5000)

for i in range(len(t)):
    randomno = np.random.choice(ran,p=[0.75,0.25])
    part = np.random.choice(m)
    time = t[i]
    
    if randomno == 1:
        if part in n:
            n.remove(part)
            n1.append(part)
    
    if randomno == 2 and len(n1)>0:
        if part in n1:
            n1.remove(part)
            n.append(part)
        
    
    plt.plot(time,len(n),marker='+',color='black',markersize='0.8')
    plt.plot(time,len(n1),marker='+',color='black',markersize='0.8')


n = np.linspace(1,200,200)
m = n
ran = np.linspace(1,2,2)
n1 = []
n = list(n)
n1 = list(n1)
    
for j in range(len(t)):
    randomno = np.random.choice(ran,p=[0.55,0.45])
    part = np.random.choice(m)
    time = t[j]
    
    if randomno == 1:
        if part in n:
            n.remove(part)
            n1.append(part)
    
    if randomno == 2 and len(n1)>0:
        if part in n1:
            n1.remove(part)
            n.append(part)
        
    
    plt.plot(time,len(n),marker='+',color='crimson',markersize='0.8')
    plt.plot(time,len(n1),marker='+',color='crimson',markersize='0.8')


red_patch = mpatches.Patch(color='crimson', label='55%/45%')
black_patch = mpatches.Patch(color='black', label='75%/25%')
 

    
plt.plot(t,hund,linestyle='-',linewidth='2')
plt.ylim([0,200])
plt.xlim([0,5000])
plt.grid()
plt.xlabel('Time step')
plt.ylabel('No. of particles at one side')
plt.title('Time evolution of box - varying probability') 
plt.legend(handles=[red_patch,black_patch])   
fig = plt.gcf()

fig.savefig('Q4b.png', dpi=100)
```
